```python
def plot_data(n):
    xval = []
    yval=[]
    for i in range(n):
        phi = (1/n*i)
        # 1 = x2 + y2 -> y = sqrt(1-x2)
        x = np.sqrt(1-phi**2)
        xval.append(x)
        yval.append(phi)

    plt.figure(figsize=[8, 8])
    plt.plot(xval, yval)

    plt.xlim(0, 1.2)
    plt.ylim(0, 1.2)
    plt.grid()

    plt.show()

interact(plot_data, n=(0, 50));
```
